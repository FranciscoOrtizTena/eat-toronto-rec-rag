```python
import requests
import time
```

```python
API_KEY = 'Your_Google_API_KEY'
```

```python
#Toronto city coordinates
low_latitude = [43.62786083370704]
low_longitude = [-79.4324836390456]
high_latitude = [43.71388464651852]
high_longitude = [-79.36912532694454]

# List of restaurant types
restaurant_types = [
    'american_restaurant',
    'bakery',
    'bar',
    'barbecue_restaurant',
    'brazilian_restaurant',
    'breakfast_restaurant',
    'brunch_restaurant',
    'cafe',
    'chinese_restaurant',
    'coffee_shop',
    'fast_food_restaurant',
    'french_restaurant',
    'greek_restaurant',
    'hamburger_restaurant',
    'ice_cream_shop',
    'indian_restaurant',
    'indonesian_restaurant',
    'italian_restaurant',
    'japanese_restaurant',
    'korean_restaurant',
    'lebanese_restaurant',
    'meal_delivery',
    'meal_takeaway',
    'mediterranean_restaurant',
    'mexican_restaurant',
    'middle_eastern_restaurant',
    'pizza_restaurant',
    'ramen_restaurant',
    'restaurant',
    'sandwich_shop',
    'seafood_restaurant',
    'spanish_restaurant',
    'steak_house',
    'sushi_restaurant',
    'thai_restaurant',
    'turkish_restaurant',
    'vegan_restaurant',
    'vegetarian_restaurant',
    'vietnamese_restaurant'
]

# Request Headers
headers = {
    'Content-Type': 'application/json',
    'X-Goog-Api-Key': API_KEY,  # Replace 'YOUR_API_KEY' with your API key
    'X-Goog-FieldMask': "places.id,places.types,places.nationalPhoneNumber,places.internationalPhoneNumber,places.formattedAddress,places.location,places.rating,places.websiteUri,places.regularOpeningHours,places.displayName,places.primaryType,places.editorialSummary,places.reviews,nextPageToken"
}

def get_restaurants(json_body, headers):
    response = requests.post('https://places.googleapis.com/v1/places:searchText', json=json_body, headers=headers)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error: {response.status_code}")
        return None

# List to store the results of all types of restaurants
all_restaurants = []

# Loop over each type of restaurant in the list
for i in range(len(low_latitude)):
    for restaurant_type in restaurant_types:
        print(f"Buscando tipo de restaurante: {restaurant_type}\n")
        print("Buscando en las coordenadas")
        print(f"low:{low_latitude[i]}, {low_longitude[i]}")
        print(f"high:{high_latitude[i]}, {high_longitude[i]}\n")
        json_body = {
            'textQuery' : restaurant_type,
            'includedType' : restaurant_type,
            'pageSize': 20,
            "locationBias": {
              "rectangle": {
                "low": {
                  "latitude": low_latitude[i],
                  "longitude": low_longitude[i]
                },
                "high": {
                  "latitude": high_latitude[i],
                  "longitude": high_longitude[i]
                }
              }
            }
        }

        # Get first page results
        restaurants = get_restaurants(json_body, headers)
        if restaurants:
            print(f"Se adicionaron {len(restaurants['places'])} lugares")
            all_restaurants.extend(restaurants.get('places', []))

        # Loop for the following pages
        while True:            
            # Wait a few seconds for the token to become valid
            time.sleep(2)
            
            if 'nextPageToken' in restaurants:
                next_page_token = restaurants.get('nextPageToken')
                json_body['pageToken'] = next_page_token
                restaurants = get_restaurants(json_body, headers)
                if 'places' in restaurants:
                    print(f"Se adicionaron {len(restaurants['places'])} lugares")
                    all_restaurants.extend(restaurants.get('places', []))
                else:
                    print("No se encontraron más lugares en 'places'.")
                    break
            else:
                print("Se adicionaron 0 lugares\n")
                break
```

```python
import json

# Specify the name of the file where you want to save the JSON
nombre_archivo = 'data_unclean.json'

# Save the JSON to file
with open(nombre_archivo, 'w') as archivo:
    json.dump(all_restaurants, archivo, indent=4)

print(f"El JSON ha sido guardado en el archivo '{nombre_archivo}'.")
```

```python
import json

# Specify the name of the file you want to open
nombre_archivo = 'data_unclean.json'

# Open the file and load the contents into a variable
with open(nombre_archivo, 'r') as archivo:
    all_restaurants = json.load(archivo)

# Now, 'all_restaurants' contains the JSON data
print(f"El JSON ha sido cargado correctamente desde '{nombre_archivo}'.")
```
